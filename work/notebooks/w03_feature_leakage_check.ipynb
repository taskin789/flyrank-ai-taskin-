## 1. Build the feature vector

I built the feature vector using the Breast Cancer dataset from scikit-learn.

The feature vector contains only predictor features. The target column was separated before modeling.

The dataset contains only numeric features, so no categorical encoding was required. Missing values were filled using the median as a general preprocessing step.

## 3. The leakage hunt

*Attack your own features: label-derived columns, future windows, product flags. Show the test.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
possible_leakage = []

keywords = ["target", "label", "result", "future", "outcome"]

for col in df.columns:
    for key in keywords:
        if key.lower() in col.lower():
            possible_leakage.append(col)

print("Possible leakage columns:", possible_leakage)
```

## Leakage Hunt

I checked the dataset for columns that could reveal the target or include future information.

The only detected column was the target column itself. It was removed from the feature vector before training.

No future information or label-derived features were observed.
