## 1. My rule and its reason codes

*Write the rule in plain words first. Then the reason codes it can output.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
from sklearn.datasets import load_breast_cancer
import pandas as pd
import os

# Load dataset
data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

# Rule-based score
df["action_score"] = df["mean radius"]

# Reason code
mean_value = df["mean radius"].mean()

df["reason_code"] = df["mean radius"].apply(
    lambda x: "HIGH_RADIUS" if x >= mean_value else "LOW_RADIUS"
)

df.head()
```

## My Rule and Its Reason Codes

I ranked records using the mean radius feature.

Records with a larger mean radius receive a higher action score because they are more likely to need attention.

Reason Codes:
- HIGH_RADIUS: Mean radius is above the dataset average.
- LOW_RADIUS: Mean radius is below the dataset average.

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Rank records
ranked_df = df.sort_values(by="action_score", ascending=False)

# Create output folder
os.makedirs("work/outputs", exist_ok=True)

# Save CSV
ranked_df.to_csv("work/outputs/baseline_action_score.csv", index=False)

print("CSV saved successfully.")
ranked_df.head()
```

## Build the Ranked Queue

The records were ranked by the action score in descending order.

The ranked queue was saved as a CSV file for later review.
