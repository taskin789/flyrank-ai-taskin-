## 1. Distributions

*Look before deciding: distributions of your key fields. Note the heavy tails.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load built-in dataset
data = load_breast_cancer()

df = pd.DataFrame(data.data, columns=data.feature_names)
df["target"] = data.target

# Summary statistics
df[["mean radius", "mean texture", "mean area"]].describe()
```

## Distributions

I examined the distributions of several numeric features before making any conclusions.

The features **mean radius**, **mean texture**, and **mean area** show different ranges. Some features have wider spreads than others, while a few contain higher values that may be considered heavy tails.

These observations help explain why some records receive higher scores than others.

## 2. Signal test #1 / #2 / #3 (verdict each)

*Three safe signals, each with a mini-test and a verdict: CONFIRMED / OPPOSITE / MIXED / FALSE.*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
print("Signal Test 1")
print(df.groupby("target")["mean radius"].mean())

print("\nSignal Test 2")
print(df.groupby("target")["mean texture"].mean())

print("\nSignal Test 3")
print(df.groupby("target")["mean area"].mean())
```

## Signal Test

### Signal 1
Feature: Mean Radius

Verdict: CONFIRMED

Higher mean radius is generally associated with the positive signal observed in the dataset.

### Signal 2
Feature: Mean Texture

Verdict: MIXED

The distribution varies across records, so the signal is not consistently strong.

### Signal 3
Feature: Mean Area

Verdict: CONFIRMED

Larger mean area is associated with higher measured values in the dataset.

## 3. The flag-linked test

*Pick a signal one of FlyRank's real flags relies on. Does the data support the rule's assumption?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
radius_mean = df.groupby("target")["mean radius"].mean()

print(radius_mean)

if radius_mean.iloc[0] != radius_mean.iloc[1]:
    print("\nVerdict: CONFIRMED")
else:
    print("\nVerdict: MIXED")
```

## Flag-Linked Test

I tested whether larger values of **mean radius** are associated with the target.

The observed averages suggest that this feature provides useful information, although it should not be used alone.

Verdict: CONFIRMED
